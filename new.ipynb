```python
import json
import  csv
```

```python
f = open('eng_data.json',encoding="utf-8")
eng = json.load(f)
f = open('punjabi_dummy.json',encoding="utf-8")
pbi = json.load(f)
```

```python
f = open('adj_final.json',encoding="utf-8")
adj = json.load(f)
f = open('adv_final.json',encoding="utf-8")
adv = json.load(f)
f = open('noun_final.json',encoding="utf-8")
noun = json.load(f)
f = open('verb_final.json',encoding="utf-8")
verb = json.load(f)
```

```python
from functools import reduce
 
def merge_dictionaries(dict1, dict2):
    merged_dict = dict1.copy()
    merged_dict.update(dict2)
    return merged_dict
 
dict_list = [adj, adv,noun,verb]

result_dict = reduce(merge_dictionaries, dict_list)

(result_dict)
```

```python
csv_file = 'dictionary_with_one_punjabi_word.csv'

# Write data to CSV file
with open(csv_file, 'w', newline='',encoding="utf-8") as file:
    writer = csv.writer(file)
    
    # Write header
    writer.writerow(['Word', 'Translation', 'Frequency'])
    
    # Write data rows
    for word, translations in result_dict.items():
        for translation, frequency in translations.items():
            writer.writerow([word, translation, frequency])
```

```python
df = pd.read_csv("dictionary_with_one_punjabi_word.csv")
```

```python
numeric_rows = df['Translation'].str.isnumeric()
df = df[~numeric_rows]
```

```python
df.to_csv('dictionary_with_one_punjabi_word.csv', index=False)
```

```python
df['English frequency']= df['Word'].apply(lambda x: eng.get(x, 0))
```

```python
df_threshold = df[(df['English frequency'] >= 8)]
```

```python
df_threshold.to_csv('dictionary_with_one_punjabi_word.csv', index=False)
```
