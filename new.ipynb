```python
import json
import  csv
```

```python
f = open('eng_data.json',encoding="utf-8")
eng = json.load(f)
f = open('punjabi_dummy.json',encoding="utf-8")
pbi = json.load(f)
```

```python
f = open('adj_final.json',encoding="utf-8")
adj = json.load(f)
f = open('adv_final.json',encoding="utf-8")
adv = json.load(f)
f = open('noun_final.json',encoding="utf-8")
noun = json.load(f)
f = open('verb_final.json',encoding="utf-8")
verb = json.load(f)
```

```python
from functools import reduce
 
def merge_dictionaries(dict1, dict2):
    merged_dict = dict1.copy()
    merged_dict.update(dict2)
    return merged_dict
 
dict_list = [adj, adv,noun,verb]

result_dict = reduce(merge_dictionaries, dict_list)

(result_dict)
```

```python
with open('final.json', 'w',encoding="utf-8") as f:
    json.dump(result_dict, f,ensure_ascii=False)
```

```python
def calculate_word_count_eng(text,eng_count):
    if text in eng_count:
        return eng_count[text]
    else:
        return 0
def calculate_word_count_punjabi(text,pbi_count):
    if text in pbi_count:
        return pbi_count[text]
    else:
        return 0

def convert_dict_to_csv(data, filename,eng_count,pbi_count):
    with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['English', 'Punjabi', 'English_Word_Count', 'Punjabi_Word_Count']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        for english, punjabi in data.items():
            english_word_count = calculate_word_count_eng(english,eng_count)
            punjabi_word_count = calculate_word_count_punjabi(punjabi,pbi_count)
            writer.writerow({'English': english, 'Punjabi': punjabi,
                             'English_Word_Count': english_word_count,
                             'Punjabi_Word_Count': punjabi_word_count})
            
convert_dict_to_csv(result_dict,"dictionary.csv",eng,pbi)
```
